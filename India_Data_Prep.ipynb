```python
census = pd.read_csv("india-districts-census-2011.csv")
```

```python
census.shape
```

```python
census.head()
```

```python
cols = ['Population','District code','District name','Male','Female','Literate','Households_with_Internet','Housholds_with_Electric_Lighting']
```

```python
census = census[cols]
```

```python
census
```

```python
census['Sex Ratio'] = round(census['Male'] / census['Female'] * 100)
```

```python
census
```

```python
census['Literacy Rate'] = round(census['Literate'] / census['Population'] *100)
```

```python
census
```

```python
final_df = latlong.merge(census , left_on='District' , right_on ='District name')
```

```python
final_df
```

```python
final_df = final_df.drop(columns='District name')
```

```python
final_df.to_csv('india.csv' , index = False)
```
